Написать функцию **fillna_with_mean**, которая заполняет пропуски в колонке target ее средними значениями по группам. Группировка происходит по колонке group.
Если в группе нет non-nan значений, заполняем target колонку значением nan.

```python
import pandas as pd

def fillna_with_mean(
    df: pd.DataFrame, target: str, group: str
) -> pd.DataFrame:
    categories = df[group].unique()
    groups = df.groupby(group, as_index=False)[target].mean()
    for cat in categories:
        df[df[group]==cat] = df[df[group]==cat].fillna(groups[groups[group]==cat][target].values[0])
    return df
```

```python
#Пример работы функции
df = pd.read_csv('data1.csv', sep=';')
df.head()
```

```python
fillna_with_mean(df, 'sales', 'category')
```

Колонки таблицы:

•	sku – SKU (Stock Keeping Unit), уникальный ID товара;

•	gmv – GMV (Gross Merchandise Volume), аналог розничного товарооборота;

•	stock – число единиц товара на складе;

•	price – цена на товар.

Формула расчёта: GMV = (цена) * (штук продано).

Необходимо написать функцию **limit_gmv** для постобработки предсказаний модели.

```python
def limit_gmv(df: pd.DataFrame) -> pd.DataFrame:
    for i, row in df.iterrows():
        if df.iloc[i]['gvm']/df.iloc[i]['price']>df.iloc[i]['stock']:
            df.at[i, 'gvm'] = df.iloc[i]['stock']*df.iloc[i]['price']
        elif df.iloc[i]['gvm']/df.iloc[i]['price']<df.iloc[i]['stock']:
            df.at[i, 'gvm'] = int(df.iloc[i]['gvm']/df.iloc[i]['price'])*df.iloc[i]['price']
    return df
```

```python
df = pd.read_csv('data2.csv', sep=';')
df.dropna(inplace=True)
df.head()
```

```python
limit_gmv(df)
```
